## Local minimization

Rosenbrock function :
$f(\mathbf{x}, a, b) = \sum_{i=1}^{N-1} a(x_{i+1} - x_i^2)^2 + (1-x_i)^2 + b.$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Rosenbrock function for 2 dimensions
def rosen(x):
    """The Rosenbrock function"""
    return (1.0 - x[0])**2 + 100.0 * (x[1] - x[0]**2)**2

# Create a grid of x and y values
x = np.linspace(-2.0, 2.0, 500)
y = np.linspace(-1.0, 3.0, 500)
X, Y = np.meshgrid(x, y)

# Calculate the Z values for the function
Z = rosen([X, Y])

# Plotting the 3D surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', rstride=1, cstride=1, antialiased=True)

# Set labels and title using LaTeX formatting
ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)
ax.set_zlabel('$f(x,y)$', fontsize=12)
ax.set_title('3D Plot of the Rosenbrock Function', fontsize=15)

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

# Save the plot
#plt.savefig('rosenbrock_3d_plot.png')

print("The plot has been generated and saved as rosenbrock_3d_plot.png.")
```

### Unconstrained minimization : Nelder-Mead Simplex algorithm

```python
import numpy as np
from scipy.optimize import minimize
```

```python
def rosen(x):
    """The Rosenbrock function"""
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)

x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])

res = minimize(rosen, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})

print(res.x)
```

```python
def rosen_with_args(x, a, b):
    """The Rosenbrock function with additional arguments"""
    return sum(a*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0) + b

x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])

res = minimize(rosen_with_args, x0, method='nelder-mead', args=(0.5, 1.), options={'xatol': 1e-8, 'disp': True})

print(res.x)
```

### Unconstrained optimization : Broyden-Fletcher-Goldfarb-Shanno algorithm 

To demonstrate this algorithm, the Rosenbrock function is again used. The gradient of the Rosenbrock function is the vector:
$$
\frac{\partial f}{\partial x_j} = \sum_{i=1}^{N} 200(x_i - x_{i-1}^2)(\delta_{i,j} - 2x_{i-1}\delta_{i-1,j}) - 2(1-x_{i-1})\delta_{i-1,j}.
$$
$$
= 200(x_j - x_{j-1}^2) - 400x_j(x_{j+1} - x_j^2) - 2(1-x_j).
$$

This expression is valid for the interior derivatives. Special cases are
$$
\frac{\partial f}{\partial x_0} = -400x_0(x_1 - x_0^2) - 2(1-x_0),
$$
$$
\frac{\partial f}{\partial x_{N-1}} = 200(x_{N-1} - x_{N-2}^2).
$$

```python
def rosen_der(x):
    xm = x[1:-1]
    xm_m1 = x[:-2]
    xm_p1 = x[2:]
    der = np.zeros_like(x)
    der[1:-1] = 200*(xm-xm_m1**2) - 400*(xm_p1 - xm**2)*xm - 2*(1-xm)
    der[0] = -400*x[0]*(x[1]-x[0]**2) - 2*(1-x[0])
    der[-1] = 200*(x[-1]-x[-2]**2)
    return der

res = minimize(rosen, x0, method='BFGS', jac=rosen_der, options={'disp': True})   

res.x
```

### Unconstrained optimization : Newton-Conjugate-Gradient algorithm

```python
def rosen_hess(x):
    x = np.asarray(x)
    H = np.diag(-400*x[:-1],1) - np.diag(400*x[:-1],-1)
    diagonal = np.zeros_like(x)
    diagonal[0] = 1200*x[0]**2-400*x[1]+2
    diagonal[-1] = 200
    diagonal[1:-1] = 202 + 1200*x[1:-1]**2 - 400*x[2:]
    H = H + np.diag(diagonal)
    return H

res = minimize(rosen, x0, method='Newton-CG', jac=rosen_der, hess=rosen_hess, options={'xtol': 1e-8, 'disp': True})    

res.x
```
