# Machine Learning and Handwritten Scripts

kraken ( https://kraken.re ) is supported with Python 3.11.9 (not with 3.14 that is currently installed on my computer). 

In case the shared library libperl.so.5.40 is missing in your sytem (as it was on mine) :
```
sudo dnf install perl-Digest-SHA perl-libs
```

Install Python virtual environment (`pyenv`)

``` 
# Install pyenv if you don’t have it yet
curl https://pyenv.run | bash   # follow the post‑install instructions

# Install a supported Python version, e.g. 3.11.9
pyenv install 3.11.9

# Create a virtual environment with that version
pyenv virtualenv 3.11.9 kraken-dev
pyenv activate kraken-dev   # or `pyenv local kraken-dev` inside the project dir
``` 

Now your terminal should look like :
```
(kraken-dev) username@fedora:~/sandbox$ 
```

Install the latest development branch (see [kraken website](https://kraken.re/main/index.html) )
```
git clone https://github.com/mittagessen/kraken.git
cd kraken
pip install --editable .

```

Download the default model for printed French (and other language ) text. This is the model to do the actual recognition of characters :

```
kraken get 10.5281/zenodo.10592716
```

Run the following command in the terminal :
```
(kraken-dev) username@fedora:~/sandbox$ kraken -i test1.png test1.txt segment -bl ocr -m catmus-print-fondue-large.mlmodel
```
