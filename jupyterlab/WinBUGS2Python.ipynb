```python
import numpy as np
N = 1000  # Population size of sparrows in the yard
p = 0.4  # Individual detection probability

# Generate 1 million binomial random variables with the given parameters
C = np.random.binomial(n=1, size=10**6, p=p)

# Calculate mean, variance and standard deviation of C
mean_C = np.mean(C)
variance_C = np.var(C)
std_dev_C = np.std(C)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Import data and calculate relevant statistics (as in previous answer)
N = 1000  # Population size of sparrows in the yard
p = 0.4  # Individual detection probability
C = np.random.binomial(n=1, size=10**6, p=p)
mean_C = np.mean(C)
variance_C = np.var(C)
std_dev_C = np.std(C)

# Create histogram with customized parameters
plt.hist(C, bins=50, color='gray', edgecolor='black', alpha=0.7, rwidth=(std_dev_C/10), density=True)

# Set labels and title for the plot
plt.title('Distribution of detected sparrows')
plt.xlabel('Sparrow count')
plt.ylabel('Frequency (normalized by population size)')

# Show the plot
plt.show()
```
