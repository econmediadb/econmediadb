# Newton-Raphson Method

Source: Saupin (2023) p.82; Robert and Casella (2004) p.19

The goal is to solve an equation of the form $f(x) = 0$. A common approach is to use a *Newton-Raphson* algorithm, which produces a sequence $x_n$ such that 
$$
x_{n+1} = x_n - \left( \left. \frac{ \partial f }{ \partial x }  \right|_{x=x_n} \right)^{-1} f(x_n)
$$
until it stabilizes around a solution of $f(x) = 0$.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
  return x**3 - 2*x - 5

# Create a list of x-values in the range [-2, 3]
x = np.linspace(-2, 3, 100)  # 100 points between -2 and 3

# Calculate the corresponding y-values using the function
y = f(x)

# Plot the function
plt.plot(x, y, label='f(x) = x^3 - 2x - 5')

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^3 - 2x - 5')

# Add a legend
plt.legend()

# Set grid
plt.grid(True)

# Show the plot
plt.show()
```

Let us linearise the following function $f(x) = x^3 -2x - 5$ at the neighbourhood of $x_0$
$$
f_{x_0}(x) = f(x_0) + f'(x_0) \times (x - x_0)
$$

```python
# Caclule numériquement la dérivée de la fonction f au point x
# $f'(x) = \lim\limits_{d\rightarrow 0} \frac{f(x+d) - f(x)}{d}$.
def deriv(f, x, d=1e-3):
  return (f(x + d) - f(x)) / d
```

```python
# Cherche une solution de f(x) = 0 au voisinage de $x_0$.
# $n$ est le nombre d'itérations.
def raphson(f, x0, n=5):
  xi = x0  # xi contient les approximations successives.
  for i in range(0, n):
    d = -f(xi) / deriv(f, xi)
    xi = xi + d
  return xi
```

$$
f(x) = x^3 - 2x -5, \ \ x\in [-2,3]
$$

```python
f = lambda x: x**3 - 2 * x - 5
r = raphson(f, 2)
print(r)
#-> 2.09455148154233
print(f(r))
# -> 3.8191672047105385e-14
```
