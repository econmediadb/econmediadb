```python
import csv
import matplotlib.pyplot as plt
import os
from datetime import datetime

# Define the path to your CSV file
csv_file_path = '../data/LU1,DF_E2501,1.0+all.csv'
#csv_file_path = '../data/testdata.csv'

# --- Data Import and Initial Inspection ---

def load_and_inspect_data_no_pandas(path):
    """
    Loads a CSV file into a list of dictionaries and displays initial information.

    Args:
        path (str): The file path to the CSV.

    Returns:
        list: A list of dictionaries, where each dictionary is a row.
    """
    if not os.path.exists(path):
        print(f"Error: The file '{path}' does not exist.")
        return None

    data = []
    try:
        with open(path, mode='r', encoding='utf-8') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                data.append(row)
        print("CSV file loaded successfully!\n")

        print("--- Data Head (first 5 rows) ---")
        for i, row in enumerate(data[:5]):
            print(row)

        print("\n--- Data Info (Column types and non-null counts) ---")
        if not data:
            print("No data to display info for.")
            return []

        # Get all unique headers
        headers = data[0].keys() if data else []
        column_info = {header: {'count': 0, 'sample_values': set()} for header in headers}

        for row in data:
            for header in headers:
                if row.get(header) not in ['', None]:
                    column_info[header]['count'] += 1
                if len(column_info[header]['sample_values']) < 5: # Collect up to 5 unique samples
                    column_info[header]['sample_values'].add(row.get(header))

        for header, info in column_info.items():
            print(f"Column: {header}")
            print(f"  Non-null count: {info['count']} / {len(data)}")
            print(f"  Sample values: {list(info['sample_values'])[:5]}") # Print up to 5 samples
            print("-" * 30)

        return data
    except Exception as e:
        print(f"Error loading CSV file: {e}")
        return None

# Load your data
data = load_and_inspect_data_no_pandas(csv_file_path)

if data is not None:
    # --- Data Cleaning and Preparation ---

    # Corrected column mapping based on actual CSV headers
    column_mapping = {
        'TIME_PERIOD': 'Time_Period',
        'OBS_VALUE': 'OBS_VALUE',
        'Labels': 'Series_Label'
    }

    cleaned_data = []
    for row in data:
        new_row = {}
        for old_key, new_key in column_mapping.items():
            new_row[new_key] = row.get(old_key)
        # Keep other original columns if needed, or explicitly copy them
        for original_key in row.keys():
            if original_key not in column_mapping and original_key not in new_row: # Avoid overwriting
                new_row[original_key] = row[original_key]

        # Convert 'Time_Period' to datetime objects for proper plotting
        time_period_str = new_row.get('Time_Period')
        if time_period_str:
            try:
                # Handle "YYYY-QX" format by converting to start of quarter
                year, quarter = time_period_str.split('-')
                quarter_num = int(quarter[1])
                # Map quarter to month and day for approximate quarter start
                if quarter_num == 1:
                    new_row['Time_Period_dt'] = datetime.strptime(f"{year}-03-31", "%Y-%m-%d")
                elif quarter_num == 2:
                    new_row['Time_Period_dt'] = datetime.strptime(f"{year}-06-30", "%Y-%m-%d")
                elif quarter_num == 3:
                    new_row['Time_Period_dt'] = datetime.strptime(f"{year}-09-30", "%Y-%m-%d")
                elif quarter_num == 4:
                    new_row['Time_Period_dt'] = datetime.strptime(f"{year}-12-31", "%Y-%m-%d")
                else:
                    new_row['Time_Period_dt'] = None # Invalid quarter
            except (ValueError, IndexError):
                new_row['Time_Period_dt'] = None # Parsing error
        else:
            new_row['Time_Period_dt'] = None

        # Convert 'OBS_VALUE' to numeric, coercing errors to None
        obs_value_str = new_row.get('OBS_VALUE')
        if obs_value_str:
            try:
                new_row['OBS_VALUE_numeric'] = float(obs_value_str)
            except ValueError:
                new_row['OBS_VALUE_numeric'] = None
        else:
            new_row['OBS_VALUE_numeric'] = None

        cleaned_data.append(new_row)

    # Drop rows where 'OBS_VALUE_numeric' or 'Time_Period_dt' is None
    cleaned_data = [
        row for row in cleaned_data
        if row.get('OBS_VALUE_numeric') is not None and row.get('Time_Period_dt') is not None
    ]

    print("\n--- Cleaned Data Info ---")
    if not cleaned_data:
        print("No cleaned data to display info for.")
    else:
        # Simplified info for cleaned data
        print(f"Total rows after cleaning: {len(cleaned_data)}")
        if cleaned_data:
            print("Sample cleaned row:")
            print({k: cleaned_data[0][k] for k in ['Time_Period', 'Time_Period_dt', 'OBS_VALUE', 'OBS_VALUE_numeric', 'Series_Label'] if k in cleaned_data[0]})


    # --- Data Analysis and Visualization ---

    # Get unique series for plotting
    unique_labels = sorted(list(set(row.get('Series_Label') for row in cleaned_data if row.get('Series_Label'))))
    print(f"\nUnique data series found: {unique_labels}")

    # Example: Plotting 'Final consumption expenditure (P3)'
    consumption_data = [
        row for row in cleaned_data
        if row.get('Series_Label') == 'Final consumption expenditure (P3)'
    ]

    if consumption_data:
        # Sort data by time period for correct plotting order
        consumption_data.sort(key=lambda x: x['Time_Period_dt'])
        time_periods = [row['Time_Period_dt'] for row in consumption_data]
        obs_values = [row['OBS_VALUE_numeric'] for row in consumption_data]

        plt.figure(figsize=(12, 6))
        plt.plot(time_periods, obs_values, marker='o', linestyle='-')
        plt.title('Final Consumption Expenditure (P3) Over Time')
        plt.xlabel('Time Period')
        plt.ylabel('Observation Value')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig('final_consumption_expenditure.png')
    else:
        print("\n'Final consumption expenditure (P3)' data not found or is empty after filtering.")

    # Example: Plotting 'Imports of services (incld Clux + cif/fob) (TRP72)'
    imports_data = [
        row for row in cleaned_data
        if row.get('Series_Label') == 'Imports of services (incld Clux + cif/fob) (TRP72)'
    ]

    if imports_data:
        # Sort data by time period for correct plotting order
        imports_data.sort(key=lambda x: x['Time_Period_dt'])
        time_periods = [row['Time_Period_dt'] for row in imports_data]
        obs_values = [row['OBS_VALUE_numeric'] for row in imports_data]

        plt.figure(figsize=(12, 6))
        plt.plot(time_periods, obs_values, marker='x', linestyle='--', color='red')
        plt.title('Imports of Services Over Time')
        plt.xlabel('Time Period')
        plt.ylabel('Observation Value')
        plt.grid(True)
        plt.tight_layout()
        plt.savefig('imports_of_services.png')
    else:
        print("\n'Imports of services (incld Clux + cif/fob) (TRP72)' data not found or is empty after filtering.")

    print("\nData loading and initial analysis complete.")
```
