# VAR Model

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.api import VAR
import matplotlib.pyplot as plt
# We no longer need load_pandas since we are using hardcoded data
# from statsmodels.tools.testing import load_pandas

# --- 1. Load User Data ---

print("--- Loading User-Provided Time Series Data (N=4) ---")
# WARNING: VAR models require significantly more data (typically 50+ observations)
# for reliable estimation. Results from N=4 are purely illustrative.

M_obs = [22, 30, 23, 48]
B_obs = [108, 140, 128, 178]
D_obs = [81, 119, 103, 124]
N = len(M_obs) 

# Combine into a Pandas DataFrame
data = pd.DataFrame({'M': M_obs, 'B': B_obs, 'D': D_obs})
print(f"Data:\n{data}\n")

# --- 1.5. Plot the Data ---

print("--- Plotting Time Series Observations ---")
# Create a line plot for each variable over the four observations
plt.figure(figsize=(10, 6))
# Using .index allows plotting against the observation number (0, 1, 2, 3)
plt.plot(data.index, data['M'], label='M', marker='o', linewidth=2)
plt.plot(data.index, data['B'], label='B', marker='s', linewidth=2)
plt.plot(data.index, data['D'], label='D', marker='^', linewidth=2)

plt.title('Time Series Observations (M, B, D)')
plt.xlabel('Observation Index')
plt.ylabel('Value')
plt.xticks(data.index) # Ensure ticks are only at the observation points
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(loc='best')
plt.tight_layout()
plt.show()
print("Plot generated successfully.\n")

# --- 2. VAR Model Estimation ---

# Instantiate the VAR model
model = VAR(data)

# Fix the lag to 1 because N=4 is too small for meaningful lag selection or higher order lags.
optimal_lag = 1 
print(f"Due to small sample size (N={N}), fixing the VAR lag to {optimal_lag} for demonstration.\n")

# Fit the VAR model using the fixed lag
results = model.fit(optimal_lag)

# --- 3. Display Results ---

print("--- VAR Model Estimation Results (Lag 1) ---")
# The summary shows the coefficients for each variable in each equation.
print(results.summary())

# NOTE: Forecasting and Causality tests (Sections 4 and 5 from the previous version) 
# have been removed as they are unreliable with only 4 data points.
```
