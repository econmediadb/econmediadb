# Ordinary Differential Equation (ODE) in Python

## Example 0a : Simple ODE

Consider the following ODE :

$ \frac{dy}{dt} = -yt + 13 $

```python
# Import libraries
import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 

# Define the differential equation
def returns_dydt(y,t): 
    dydt = -y * t + 13
    return dydt 
  
# Set up initial condition 
y0 = 1
  
# Choose values of time 
t = np.linspace(0,5) 
  
# Solving the ODE 
y = odeint(returns_dydt, y0, t) 
  
# Plotting the Results by using Matplotlib 
plt.plot(t,y) 
plt.xlabel("Time") 
plt.ylabel("Y") 
plt.show()
```

$ \frac{dy}{dt} = \frac{ 1-y }{ 1.95-y } - \frac{ y }{ 0.05 + y } $

```python
# Import libraries
import numpy as np 
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 

# Define the differential equation
def returns_dydt(y,t): 
	dydt = (1-y)/(1.95-y) - y/(0.05+y) 
	return dydt 

# Set up initial condition 
y0 = [0, 1, 2] 

# Choose values of time 
t = np.linspace(1,10) 

# Solving the ODE 
y = odeint(returns_dydt, y0, t) 

# Plotting the Results by using Matplotlib 
plt.plot(t,y) 
plt.xlabel("Time") 
plt.ylabel("Y") 
plt.show()
```

## Example 1 : Exponential decay

An ordinary differential equation is an equation that relates a function to its derivatives. A simple example is:

$[ \frac{dy}{dt} = -ky ]$

where ( $k$ ) is a constant, $t$ is the independent variable (time), and $y$ is the dependent variable (the quantity that is decaying). This equation describes [exponential decay](https://en.wikipedia.org/wiki/Exponential_decay).



```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Define the differential equation
def decay(t, y, k):
    return -k * y
```

```python
# Set up initial conditions
y0 = [10] # Initial condition
k = 0.3 # Decay constant
```

The previous lines set our initial conditions. Suppose we start with an initial quantity of ( $y_0 = 10$ ), and let’s choose ( $k = 0.3$ ):

```python
# Choosing the time
t_span = (0, 20) # Start and end time
t_eval = np.linspace(t_span[0], t_span[1], 100) # Time points to evaluate the solution
```

The previous lines define the time over which we want to solve the ODE. Let’s say we want to observe the decay from ( $t = 0$ ) to ( $t = 20$ ).

```python
# Solving the Differential Equation
solution = solve_ivp(decay, t_span, y0, args=(k,), t_eval=t_eval)
```

By using the ´solve_ivp´ function from SciPy we will solve the ODE. We will pass our decay function, time, initial conditions, and any additional arguments (like ( $k$ )).

```python
# Analyzing the Results
t = solution.t
y = solution.y[0]
```

The solution object contains the results of our integration. We can extract the time points and the corresponding values of ( $y$ ).

```python
# Plotting the Results by using Matplotlib
plt.plot(t, y)
plt.title('Exponential Decay')
plt.xlabel('Time')
plt.ylabel('Quantity')
plt.grid()
plt.show()
```

The above graph shows how the quantity decreases over time.

## Example 2 : Modeling Population Growth

We can also use an ODE to model the growth of a population using the logistic growth model.

Define the Logistic Growth Model: The logistic growth can be described by the equation:  

$ \frac{dP}{dt} = rP \left(1 — \frac{P}{K}\right) $ 

where ( $P$ ) is the population size, ( $r$ ) is the growth rate, and ( $K$ ) is the carrying capacity.

```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the Logistic Growth Model
def logistic_growth(P, t, r, K):
    dPdt = r * P * (1 - P / K)
    return dPdt

# Set Initial Conditions and Parameters
P0 = 10 # initial population
r = 0.1 # growth rate
K = 100 # carrying capacity
t = np.linspace(0, 100, 100) # time from 0 to 100

# Solve the ODE
P = odeint(logistic_growth, P0, t, args=(r, K))

# Plot the Results
plt.plot(t, P)
plt.title('Logistic Growth of Population')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.grid()
plt.show()
```

The above diagram shows a curve that starts with a rapid increase in population and then levels off as it approaches the carrying capacity.

## Example 3 : Simulating a Simple Harmonic Oscillator

We can use ODE to model the motion of a simple harmonic oscillator using differential equations.

The motion can be described by the equations: 

$\frac{dx}{dt} = v$ 

$\frac{dv}{dt} = -\omega^2 x$ 

where ( $x$ ) is the position, ( $v$ ) is the velocity, and ( $\omega$ ) is the angular frequency.

```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the System of Equations
def harmonic_oscillator(Y, t, omega):
    x, v = Y
    dydt = [v, -omega**2 * x]
    return dydt

# Set Initial Conditions and Parameters
omega = 2 * np.pi # frequency
Y0 = [1.0, 0.0] # initial position and velocity
t = np.linspace(0, 10, 100) # time from 0 to 10 seconds

# Solve the ODE
Y = odeint(harmonic_oscillator, Y0, t, args=(omega,))

# Plot the Results
plt.plot(t, Y[:, 0])
plt.title('Simple Harmonic Oscillator')
plt.xlabel('Time (s)')
plt.ylabel(‘Position (m)’)
plt.grid()
plt.show()
```

The above diagram shows a sinusoidal wave representing the oscillation of the mass over time.

## Example 4 : The SIR Model in Epidemiology

We can use ODEs to model the spread of an infectious disease using the SIR model.


The SIR model is defined by: 

$ \frac{dS}{dt} = -\beta \frac{S I}{N} $ 

$ \frac{dI}{dt} = \beta \frac{S I}{N} — \gamma I $ 

$ \frac{dR}{dt} = \gamma I $ 

where ( $S$ ) is susceptible, ( $I$ ) is infected, ( $R$ ) is recovered, ( $N$ ) is the total population, ( $\beta$ ) is the transmission rate, and ( $\gamma$ ) is the recovery rate.

```python
# Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the SIR Model
def sir_model(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Set Initial Conditions and Parameters  
N = 1000 # total population
I0 = 1 # initial infected
R0 = 0 # initial recovered
S0 = N - I0 - R0 # initial susceptible
beta = 0.3 # transmission rate
gamma = 0.1 # recovery rate
t = np.linspace(0, 160, 160) # time from 0 to 160 days
y0 = S0, I0, R0

# Solve the ODE
ret = odeint(sir_model, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the Results
plt.plot(t, S, 'b', label='Susceptible')
plt.plot(t, I, 'r', label='Infected')
plt.plot(t, R, 'g', label='Recovered')
plt.title('SIR Model of Infectiuous Disease')
plt.legend(['Susceptible', 'Infected', 'Recovered'])
```
