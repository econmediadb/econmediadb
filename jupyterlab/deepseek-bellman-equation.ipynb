### Bellman Equation for RBC Model

The value function $ V(k, a) $ satisfies:

$$
V(k, a) = \max_{c, l} \left[ U(c, 1 - l) + \beta E_t [V(k', a')] \right]
$$

**Subject to the constraint:**

$$
k' = (1 - \delta)k + a f(k, l) - c
$$

Where:
- $ k $ is current capital.
- $ a $ is the productivity shock.
- $ c $ is consumption.
- $ l $ is labor supply.
- $ k' $ is next period's capital.
- $ U(c, 1 - l) $ is utility from consumption and leisure.
- $ f(k, l) $ is the production function.
- $ \delta $ is the depreciation rate.
- $ \beta $ is the discount factor.

This equation encapsulates the household's optimal decisions regarding consumption and labor, considering future capital accumulation and productivity shocks.
